```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
```

```python
dframe = pd.read_csv('marketing_customer_analysis.csv')
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
```

```python
dframe.head()
```

Step 1: Split X-y based on the original dataset

```python
#Create a copy of the original dataframe
dframe_copy = dframe.copy()
```

```python
y = dframe_copy['Total Claim Amount']
X = dframe_copy.drop(['Total Claim Amount'], axis=1)
```

```python
y.head()
```

```python
X.head()
```

Step 2: Split X values into two categories: Num and Cat

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = np.object)
```

```python
X_num.head()
```

```python
X_cat.head()
```

Step 3: Apply normalization and onehotencoder to the X values

```python
#Normalization
transformer = Normalizer().fit(X_num)
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
```

```python
x_normalized
```

```python
df_normalized = pd.DataFrame(x_normalized, columns = X_num.columns)
```

```python
df_normalized
```

```python
#Remove unnecessary columns
from sklearn.preprocessing import OneHotEncoder
X_cat.drop(['Customer', 'Effective To Date'], axis=1, inplace = True)
```

```python
X_cat.head()
```

```python
X_cat.columns
```

```python
# Apply onehotencoder to X_Cat
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(X_cat)
enc.categories_
```

```python
enc.transform(X_cat).toarray()
```

```python
column_name = enc.get_feature_names(['State', 'Response', 'Coverage', 'Education', 'EmploymentStatus',
       'Gender', 'Location Code', 'Marital Status', 'Policy Type', 'Policy',
       'Renew Offer Type', 'Sales Channel', 'Vehicle Class', 'Vehicle Size'])
```

```python
df_enc=pd.DataFrame(enc.transform(X_cat).toarray(), columns = column_name)
#df_cat.columns = X_cat.columns
df_enc.head()
```

```python
df_concat = pd.concat([df_normalized,df_enc],axis = 1)
```

```python
df_concat.head()
```

Step 4: Train-test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_concat,y, test_size = 0.2)
```

Step 5: Fit model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

# We "train" our linear model based on the data available on X_train and y_train
model = lm.fit(X_train,y_train)
```

Step 6: Prediction and measure the accuracy

```python
predictions = lm.predict(X_test)
```

```python
#model_score=model.score(X_train, y_train)
from sklearn.metrics import mean_squared_error, r2_score

print("The R2 of the linear model is: ",round(r2_score(y_test, predictions),3))
```

MSE

```python
mse = mean_squared_error(y_test, predictions)
print(mse)
```

RMSE

```python
import math 
from math import sqrt
rmse = math.sqrt(mse)
print(rmse)
```
